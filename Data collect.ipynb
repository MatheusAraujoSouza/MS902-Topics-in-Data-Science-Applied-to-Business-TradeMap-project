```
def _findTime(siteStr, index):
    """

    :param siteStr:
    :param index:
    :return:
    """

    website = requests.get(siteStr)
    soup = BeautifulSoup(website.content, 'html.parser')

    if (index == 1):

        vec = soup.find_all("time")
        tag = vec[-1]
        tag_ = tag.attrs
        tag_["datetime"] = tag_["datetime"].replace("T", " ")

        return tag_["datetime"]

    elif (index == 8):

        soup = BeautifulSoup(website.content, 'html.parser')

        vec = soup.find_all("script")

        tag = vec[0].contents
        tagStr = tag[0]
        index = tagStr.find("2021")
        rest = tagStr[index:index + 19]

        rest = rest.replace("T", " ")

        return rest

    else:

        soup = BeautifulSoup(website.content, 'html.parser')

        vec = soup.find_all("meta", attrs = {"property": "article:published_time"})
        tag = vec[0]
        tag_ = tag.attrs
        tag_["content"] = tag_["content"].replace("T", " ")

        return tag_["content"]
```

```
def removeDuplicate(fileName):
    """
    Função que remove as linhas repetidas
    do arquivo .csv final

    :param fileName: Nome do arquivo .csv
    :return: Nothing
    """

    df = pd.read_csv(fileName, sep = ",")
    df.drop_duplicates(subset = ["Link", "Empresa"], inplace = True)
    df.to_csv(fileName, index = False)
```

```
def _findTheString(tag, findStr):
    """
    Função que verifica se a string está em outra em outro texto
    Isso é feito procurando a string tag dentro de outra string

    Em nosso caso, consideramos que a string tag deve aparecer exatamente
    do jeito que ela está escrito e não consideramos os casos de
    sub-strings durante a busca

    :param tag: String que fará a busca
    :param findStr: String que será percorrida para veriricar a tag
    :return: retorna 1 se encontrou a string e 0 caso contrário
    """

    t = split(r"\s", findStr)
    tag = split(r"\s", tag)

    lenT = len(t)
    lenTag = len(tag)

    for i in range(0, lenT):

        if (tag[0] == t[i]):

            for j in range(1, lenTag):

                if ((i + 1) < lenT and tag[j] != t[i + 1] or (i + 1) >= lenT):

                    return 0

                i += 1

            return 1

    return 0
```

```
def APIRequests():
    """
    Função que realiza a requisição dos
    requests para o programa
    Além de salvar em um aquivo .CSV do que precisa

    :return: Nothing
    """

    df= spark.sql("select * from tabela_final_internacional_csv").toPandas()

    # Base de dados da notícia
    sitesStr = ["https://oglobo.globo.com", "https://www.infomoney.com.br", "https://valor.globo.com",
                "https://comoinvestir.thecap.com.br", "https://br.investing.com",
                "https://www.seudinheiro.com", "https://www.ultimoinstante.com.br", "https://www.criptofacil.com",
                "https://livecoins.com.br", "https://braziljournal.com", "https://beincrypto.com.br"]

    empresas = df["Empresa"]
    setor = df["Setor econômico"]
    ativo_ = df["Listagem"]

    # for para percorrer a tabela do dataset
    for k in range(df.shape[0]):

        tag = empresas[k]
        ativo = ativo_[k]
        setorEco = setor[k]

        tamSites = len(sitesStr)
        csvFinal = [[]]

        # Será as colunas do arquivo .CSV
        header = ["Link", "Notícia", "Empresa", "Setor Econômico", "Data de noticia"]

        for i in range(0, tamSites):

            # Using the requests module, we use the "get" function
            # provided to access the webpage provided as an
            # argument to this function:
            results = requests.get(sitesStr[i], verify = True)

            # To make sure that the website is accessible, we can
            # ensure that we obtain a 200 OK response to indicate
            # that the page is indeed present:
            # print(results.status_code)

            # For other potential status codes you may encounter,
            # consult the following Wikipedia page:
            # https://en.wikipedia.org/wiki/List_of_HTTP_status_codes

            # We can also check the HTTP header of the website to
            # verify that we have indeed accessed the correct page:
            # print(results.headers)

            # For more information on HTTP headers and the information
            # one can obtain from them, you may consult:
            # https://en.wikipedia.org/wiki/List_of_HTTP_header_fields

            # Now, let us store the page content of the website accessed
            # from requests to a variable:
            src = results.content

            # Now that we have the page source stored, we will use the
            # BeautifulSoup module to parse and process the source.
            # To do so, we create a BeautifulSoup object based on the
            # source variable we created above:
            soup = BeautifulSoup(src, 'lxml')

            # Now that the page source has been processed via Beautifulsoup
            # we can access specific information directly from it. For instance,
            # say we want to see a list of all of the links on the page:
            links = soup.find_all("a")

            print("Procurando no site {}\n".format(sitesStr[i]))
            linkNoticia = []
            tituloNoticia = []
            dataNot = []
            empresa = tag
            tamNot = 0

            """
            Para a busca, estamos realizando uma busca em casos maiúsculo,
            minúsculo e capitalizado (primeira letra maúscula)
            
            """

            for link in links:

                findStr = unidecode.unidecode(link.text.upper())

                if (_findTheString(tag, findStr) == 1 and "title" in link.attrs or
                        (ativo != "" and _findTheString(ativo, findStr) == 1 and "title" in link.attrs)):

                    tamNot += 1
                    now = datetime.datetime.now(pytz.timezone('America/Sao_Paulo'))
                    linkNoticia.append(link.attrs['href'])
                    tituloNoticia.append(link.attrs['title'])
                    dataNot.append(now)

            if (tamNot == 0):

                print("Não foi encontrado nenhuma notícia relacionado ao tema {} no site {} !!!\n".format(tag, sitesStr[i]))

            else:

                print("Achamos {} notícias de seu interesse no site {}\n".format(tamNot, sitesStr[i]))

                for j in range(0, tamNot):

                    print("\n---------- Notícia {} ----------".format(j + 1))
                    print("Título da notícia : {}\n".format(tituloNoticia[j]))
                    print("Link da notícia : {}\n".format(linkNoticia[j]))
                    print("Nome da empresa : {}\n".format(empresa))
                    print("Setor econômico da empresa : {}\n".format(setorEco))
                    print("Data da notícia : {}\n".format(str(dataNot[j].day) + "/" + str(dataNot[j].month) + "/" + str(dataNot[j].year)))
                    print("Hora da notícia : {}\n".format(str(dataNot[j].hour) + " : " + str(dataNot[j].minute) + " : " + str(dataNot[j].second)))
                    print("--------------------------------\n")

                    # Caso em que ela começa vazia
                    if(not csvFinal[0]):

                        csvFinal[0] = [linkNoticia[j], tituloNoticia[j], empresa, setorEco, dataNot[j]]

                    else:

                        csvFinal.append([linkNoticia[j], tituloNoticia[j], empresa, setorEco, dataNot[j]])


        """ Precisa implementar nessa parte um csv que reescreve
            o Arquivo .csv a partir do nome do setor
            Mas ele precisa verificar se o csv já existe
            Caso ele exista, ele só sobre-reescreve ele"""
        if (len(csvFinal[0]) > 0):

            # Essa parte verifica se o .csv NÃO existe
            # Caso ele não existe, ele entra nessa função e cria o arquivo
            if (os.path.isfile("DadosColeta8.csv") == False):

                # open the file in the write mode
                with open("DadosColeta8.csv", "a", encoding = "UTF8") as f:
                    # create the csv writer
                    writer = csv.writer(f)

                    # write a row to the csv file
                    writer.writerow(header)
                    writer.writerows(csvFinal)

                    f.close()

            # Nesta parte, ele só junta o arquivo, caso ele exista
            else:

                # open the file in the write mode
                with open("DadosColeta8.csv", "a", encoding="UTF8") as f:
                    # create the csv writer
                    writer = csv.writer(f)

                    # write a row to the csv file
                    writer.writerows(csvFinal)

                    f.close()
```

```
#Funções para obtenção e classificação de notícias de acordo com um input de empresa:

def news_printing(company):
    
    sector = find_sector(company) #Find the economic sector of the desired company
    
    news = make_news_list(sector, company) #Makes the table with news to be printed
    
    news = calculate_relevance(news, company) #Reorder the news table with respect to relevance criteria
    
    print_news(news) #Print the news according to relevance
        
    return news


def find_sector(company): #Find the sector of the desired company
        
    setores = spark.sql("select * from tabela_formatada2__csv").toPandas() #Databricks format, obtain sectors data frame
        
    setores = setores.iloc[:,0:5]
    
    for i in range(setores.shape[0]):
        if company == setores.iloc[i,0]:
            sector = setores.iloc[i,1] #Determiness the sector of the chosen company
                    
    return sector

def make_news_list(sector, company): #Makes the table with news to be printed

    all_news = pd.read_csv("DadosColeta8.csv") #Extract news from table
   
    news = pd.DataFrame(columns = all_news.columns) #Initializes a data frame which is going to contain news to be plotted
    
    for i in range(all_news.shape[0]): #For each row
        if all_news.iloc[i,2] == company: #If finds a new from the company
            news = news.append(all_news.iloc[i,:]) #Add company news to data frame
            
    if news.shape[0]<2: #If there are less than 2 news of the desired company, adds other news from sector
        
        for i in range(all_news.shape[0]): #For each row, from last new to newer:
            if all_news.iloc[-i-1,2] != company and all_news.iloc[-i-1,3] == sector: #If finds a new that is not from the company but is from the same sector
                news = news.append(all_news.iloc[-i-1,:]) #Add company news to data frame
            if news.shape[0]>=4 : #Stops when there are 4 news in total in the data frame
                break
        
    return news

def calculate_relevance(news, company):
    #Relevance function based on time from news adding to data frame
    #OBS: max of 10 days of news storage = 864000 seconds
    
    current_time = datetime.datetime.now(pytz.timezone('America/Sao_Paulo')) #Current time
    
    news["Delta_t"] = 0 #"Age" of news in seconds
    
    #Calculating delta_time from new adding to data frame
    for i in range(news.shape[0]): #For each new in the data frame,
        new_age = datetime.datetime.strptime(news.iloc[i,4], '%Y-%m-%d %H:%M:%S.%f%z') #Calculate news age
        delta_t = (current_time - new_age).total_seconds() #Calculate time from news reading to now
        news.iloc[i,5] = delta_t #Adds delta time from news reading to now to the data frame
        
    #Calculating relevance:  
    max_days = 10 #Max days for which the news are going to stay in data frame 
    max_seconds = max_days*60*60*24 #Max seconds for which the news are going to stay in data frame 
    
    relevance = 1 - (news["Delta_t"]/(max_seconds)) #Relevance = 1 if new has just been added,
                                                    #and = 0 if it's been added 10 days ago
    news["Relevância"] = relevance #Creates a new column in data frame
    
    news = news.sort_values("Relevância", ascending = False) #Sort news by relevance
        
    return news
  

def delta_t(time):
    #Calculates how many days, hours, minutes and seconds are within a number of seconds
    day = time // (24 * 3600)
    time = time % (24 * 3600)
    hour = time // 3600
    time %= 3600
    minutes = time // 60
    time %= 60
    seconds = time
    return day, hour, minutes, seconds

def print_news(news):
    
    if news.shape[0] == 1:
      print("Esta notícia pode te interessar:")
    else:
      print("Estas notícias podem te interessar:")
    print("====================================================================================================")
    
    if news.shape[0]==0: #In case 0 news are found:
      print("Nos últimos 10 dias, nenhuma notícia foi registrada sobre essa empresa, nem sobre empresas relacionadas.")
    else:
      for i in range(news.shape[0]):  

          #Day and hour of news:
          data = datetime.datetime.strptime(news.iloc[i,4], '%Y-%m-%d %H:%M:%S.%f%z')
          dia = data.date().strftime('%d/%m/%Y')
          hora = data.time().strftime('%H:%M:%S')

          days, hours, minutes, seconds = delta_t(news.iloc[i,5]) #Time since news obtaining

          #Print news in the data frame
          print("Notícia: " + news.iloc[i,1])
          print("Link: " + news.iloc[i,0])
          print("Obtida no dia: " + dia + " às " + hora)
          print("Tempo desde a notícia: %d dias, %d horas e %d minutos." % (days, hours, minutes))

          if i < news.shape[0]-1: print("----------------------------------------------------------------------------------------------------")
        
    print("====================================================================================================")      
    
    
    return
```
