```
#Funções para obtenção e classificação de notícias de acordo com um input de empresa:

def news_printing(company):
    
    sector = find_sector(company) #Find the economic sector of the desired company
    
    news = make_news_list(sector, company) #Makes the table with news to be printed
    
    news = calculate_relevance(news, company) #Reorder the news table with respect to relevance criteria
    
    print_news(news) #Print the news according to relevance
        
    return news


def find_sector(company): #Find the sector of the desired company
        
    setores = spark.sql("select * from tabela_final_internacional_csv").toPandas() #Databricks format, obtain sectors data frame
        
    setores = setores.iloc[:,0:5]
    
    for i in range(setores.shape[0]):
        if company == setores.iloc[i,0]:
            sector = setores.iloc[i,1] #Determiness the sector of the chosen company
                    
    return sector

def make_news_list(sector, company): #Makes the table with news to be printed

    all_news = pd.read_csv("DadosColeta8.csv") #Extract news from table
   
    news = pd.DataFrame(columns = all_news.columns) #Initializes a data frame which is going to contain news to be plotted
    
    for i in range(all_news.shape[0]): #For each row
        if all_news.iloc[i,2] == company: #If finds a new from the company
            news = news.append(all_news.iloc[i,:]) #Add company news to data frame
            
    if news.shape[0]<2: #If there are less than 2 news of the desired company, adds other news from sector
        
        for i in range(all_news.shape[0]): #For each row, from last new to newer:
            if all_news.iloc[-i-1,2] != company and all_news.iloc[-i-1,3] == sector: #If finds a new that is not from the company but is from the same sector
                news = news.append(all_news.iloc[-i-1,:]) #Add company news to data frame
            if news.shape[0]>=4 : #Stops when there are 4 news in total in the data frame
                break
        
    return news

def calculate_relevance(news, company):
    #Relevance function based on time from news adding to data frame
    #OBS: max of 10 days of news storage = 864000 seconds
    
    current_time = datetime.datetime.now(pytz.timezone('America/Sao_Paulo')) #Current time
    
    news["Delta_t"] = 0 #"Age" of news in seconds
    
    #Calculating delta_time from new adding to data frame
    for i in range(news.shape[0]): #For each new in the data frame,
        new_age = datetime.datetime.strptime(news.iloc[i,4], '%Y-%m-%d %H:%M:%S.%f%z') #Calculate news age
        delta_t = (current_time - new_age).total_seconds() #Calculate time from news reading to now
        news.iloc[i,5] = delta_t #Adds delta time from news reading to now to the data frame
        
    #Calculating relevance:  
    max_days = 10 #Max days for which the news are going to stay in data frame 
    max_seconds = max_days*60*60*24 #Max seconds for which the news are going to stay in data frame 
    
    relevance = 1 - (news["Delta_t"]/(max_seconds)) #Relevance = 1 if new has just been added,
                                                    #and = 0 if it's been added 10 days ago
    news["Relevância"] = relevance #Creates a new column in data frame
    
    news = news.sort_values("Relevância", ascending = False) #Sort news by relevance
        
    return news
  

def delta_t(time):
    #Calculates how many days, hours, minutes and seconds are within a number of seconds
    day = time // (24 * 3600)
    time = time % (24 * 3600)
    hour = time // 3600
    time %= 3600
    minutes = time // 60
    time %= 60
    seconds = time
    return day, hour, minutes, seconds

def print_news(news):
    
    if news.shape[0] == 1:
      print("Esta notícia pode te interessar:")
    else:
      print("Estas notícias podem te interessar:")
    print("====================================================================================================")
    
    if news.shape[0]==0: #In case 0 news are found:
      print("Nos últimos 10 dias, nenhuma notícia foi registrada sobre essa empresa, nem sobre empresas relacionadas.")
    else:
      for i in range(news.shape[0]):  

          #Day and hour of news:
          data = datetime.datetime.strptime(news.iloc[i,4], '%Y-%m-%d %H:%M:%S.%f%z')
          dia = data.date().strftime('%d/%m/%Y')
          hora = data.time().strftime('%H:%M:%S')

          days, hours, minutes, seconds = delta_t(news.iloc[i,5]) #Time since news obtaining

          #Print news in the data frame
          print("Notícia: " + news.iloc[i,1])
          print("Link: " + news.iloc[i,0])
          print("Obtida no dia: " + dia + " às " + hora)
          print("Tempo desde a notícia: %d dias, %d horas e %d minutos." % (days, hours, minutes))

          if i < news.shape[0]-1: print("----------------------------------------------------------------------------------------------------")
        
    print("====================================================================================================")      
    
    
    return
```
